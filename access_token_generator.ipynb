```python
class Trader:
    def __init__(self, index, expiry_date, access_token, fny, pnl_from_date, pnl_to_date, segment, initial_pnl):
        self.instrument_key_map_dict = {'NIFTY':'Nifty 50', 'BANKNIFTY': 'Nifty Bank'}
        self.instrument_index = 'NSE_INDEX'
        self.index_lot_size = {'NIFTY': 25, 'BANKNIFTY':15}
        self.map_transaction_type = {"create": "SELL", "close": "BUY"}
        self.index = index
        self.expiry_date = expiry_date
        self.min_thresold = 0.40
        self.max_thresold = 0.55
        self.atm_delta = 0.50
        self.delta_breach_thresold = 0.10
        self.access_token = access_token
        self.headers = {'Accept': 'application/json', 'Content-Type': 'application/json', 'Authorization': f'Bearer {self.access_token}'}
        self.url_option_chain = 'https://api.upstox.com/v2/option/chain'
        self.url_positions = 'https://api.upstox.com/v2/portfolio/short-term-positions'
        self.url_place_position = "https://api-hft.upstox.com/v2/order/place"
        self.url_instrument_token = "https://assets.upstox.com/market-quote/instruments/exchange/NSE.csv.gz"
        self.url_pnl = 'https://api.upstox.com/v2/trade/profit-loss/data'
        self.url_meta_pnl = 'https://api.upstox.com/v2/trade/profit-loss/metadata'
        self.instrument_token_zip_fl_nm = "nse_instruments.csv.gz"
        self.instrument_token_csv_fl_nm = "nse_instruments.csv"
        self.map_order_type = {"create": "SELL", "close": "BUY"}
        self.fny = fny
        self.pnl_from_date = pnl_from_date
        self.pnl_to_date = pnl_to_date
        self.segment = segment
        self.start_bold = "\u001b[1m"
        self.end_bold = "\033[0m"
        self.instrument_token_df = self.download_instrument_token()
        self.instrument_trading_symbol_map_dict = self.instrument_token_trading_symbol_map()
        self.old_pnl = self.fetch_pnl()

    def instrument_token_trading_symbol_map(self):
        '''This method creates a mapping dict of instrument token & trading symbol'''
        temp_map_dict = dict()
        instrument_token_df_exp = self.instrument_token_df[self.instrument_token_df.expiry == self.expiry_date]
        for _, row in instrument_token_df_exp.iterrows():
            temp_map_dict[row['instrument_key']] = row['tradingsymbol']
        return temp_map_dict


    def fetch_pnl(self, pnl_from_date=None, pnl_to_date=None, segment=None, fny=None):
        '''Fetch the pnl for a given segement, for a given dater range & financical year'''
        try:
            if pnl_from_date is None or pnl_to_date is None or segment is None or fny is None:
                params = {
                    'from_date': self.pnl_from_date,
                    'to_date': self.pnl_to_date,
                    'segment': self.segment,
                    'financial_year': self.fny,
                    'page_number': '1',
                    'page_size': '5000'
                    }
            else:
                params = {
                    'from_date': pnl_from_date,
                    'to_date': pnl_to_date,
                    'segment': segment,
                    'financial_year': fny,
                    'page_number': '1',
                    'page_size': '5000'
                    }


            response = requests.get(self.url_pnl, headers=self.headers, params=params)
            if response.status_code == 200 and response.json()['data']:
                pnl_df = pd.DataFrame(response.json()['data'])
                pnl_df['pnl'] = pnl_df['sell_amount'] -  pnl_df['buy_amount']
                return int(pnl_df['pnl'].sum())
            else:
                return 0.0
        except Exception as e:
            print(f"all_pnl: Failed to fetch pnl{e}")
            traceback.print_exc()

    def check_delta_breach(self):
        '''Checks the delta gap between the active legs & take the necessory actions.
        '''
        try:
            # Get the current positions along with theirs delta
            get_active_pos_geeks_status, tot_pnl, act_pos_delta_max, act_pos_delta_min = self.get_active_pos_geeks()
            if not get_active_pos_geeks_status:
                print(f"Failed to fetch active positions with geeks")
                return False
            # delta breach logic
            current_delta_diff = abs(act_pos_delta_max['delta'] - act_pos_delta_min['delta'])
            if current_delta_diff > self.delta_breach_thresold:
                print("Delta thresold is breached. Taking the necessory actions")
                delta_list = [act_pos_delta_max['delta'], act_pos_delta_min['delta']]
                delta_list = [delta for delta in delta_list if delta>self.min_thresold and delta<self.max_thresold]
                if not delta_list:  # Both the legs delta is out of range
                    print("Both legs have breached the delta threshold. Hence, both are being closed.")
                    create_close_leg_status = self.create_close_leg(act_pos_delta_max, typ='close')
                    if not create_close_leg_status:
                        print(f"Failed to close {act_pos_delta_max['tradingsymbol']}")
                        return False
                    print(f"{act_pos_delta_max['tradingsymbol']} has been closed successfully!")
                    create_close_leg_status = self.create_close_leg(act_pos_delta_min, typ='close')
                    if not create_close_leg_status:
                        print(f"Failed to close {act_pos_delta_min['tradingsymbol']}")
                        return False
                    print(f"{act_pos_delta_min['tradingsymbol']} has been closed successfully!")
                    # Collecting required details to place a position
                    get_instrument_token_status1, instrument_token1, tradingsymbol1 = self.get_instrument_token(option_type='CE', delta=self.atm_delta)
                    if not get_instrument_token_status1:
                        print(f"Failed to get the instrument token for new CE position with delta {self.atm_delta}")
                        return False
                    get_instrument_token_status2, instrument_token2, tradingsymbol2 = self.get_instrument_token(option_type='PE', delta=self.atm_delta)
                    if not get_instrument_token_status2:
                        print(f"Failed to get the instrument token for new PE position with delta {self.atm_delta}")
                        return False
                    pos1 = {"instrument_token": instrument_token1, "quantity": act_pos_delta_max["quantity"], "tradingsymbol": tradingsymbol1}
                    pos2 = {"instrument_token": instrument_token2, "quantity": act_pos_delta_max["quantity"], "tradingsymbol": tradingsymbol2}
                    # create leg
                    create_close_leg_status = self.create_close_leg(pos1, typ='create')
                    if not create_close_leg_status:
                        print(f"Failed to create {pos1['tradingsymbol']}!")
                        return False
                    print(f"{pos1['tradingsymbol']} has been created successfully!")
                    create_close_leg_status = self.create_close_leg(pos2, typ='create')
                    if not create_close_leg_status:
                        print(f"Failed to create {pos2['tradingsymbol']}!")
                        return False
                    print(f"{pos2['tradingsymbol']} has been created successfully!")
                    return True
                else:
                    max_delta = max(delta_list)
                    if max_delta == act_pos_delta_max['delta']:
                        holding_position = act_pos_delta_max['delta']
                        closing_position = act_pos_delta_min['delta']
                    elif max_delta == act_pos_delta_min['delta']:
                        holding_position = act_pos_delta_min['delta']
                        closing_position = act_pos_delta_max['delta']
                    # Close the position
                    create_close_leg_status = self.create_close_leg(closing_position, typ='close')
                    if not create_close_leg_status:
                        print(f"Failed to close {closing_position['tradingsymbol']}")
                        return False
                    print(f"{closing_position['tradingsymbol']} has been closed successfully!")
                    # create new position dict
                    get_instrument_token_status, instrument_token, tradingsymbol = self.get_instrument_token(option_type=closing_position['option_type'], delta=holding_position['delta'])
                    if not get_instrument_token_status:
                        print(f"Failed to get the instrument token for new position of type {closing_position['option_type']} with delta {holding_position['delta']}")
                        return False
                    new_position = {"instrument_token": instrument_token, "quantity": act_pos_delta_max["quantity"], "tradingsymbol": tradingsymbol}
                    # create position
                    create_close_leg_status = self.create_close_leg(new_position, typ='create')
                    if not create_close_leg_status:
                        print(f"Failed to create {new_position['tradingsymbol']}")
                        return False
                    print(f"{new_position['tradingsymbol']} has been created successfully!")
                    return True
            else:
                tot_pnl += self.old_pnl
                print(f'\n{self.start_bold}-------------{datetime.datetime.now().strftime("%Y-%m-%d:%H:%M:%S")}---------------{self.end_bold}')
                print(f"{act_pos_delta_max['tradingsymbol']} - delta:{self.start_bold}{act_pos_delta_max['delta']}{self.end_bold} ltp:{self.start_bold}{act_pos_delta_max['last_price']}{self.end_bold}")
                print(f"{act_pos_delta_min['tradingsymbol']} - delta:{self.start_bold}{act_pos_delta_min['delta']}{self.end_bold} ltp:{self.start_bold}{act_pos_delta_min['last_price']}{self.end_bold}")
                print(f"Delta diff: {self.start_bold} {round(current_delta_diff, 7)}{self.end_bold}")
                print(f"PNL: {self.start_bold} {round(tot_pnl)}{self.end_bold}")
                print(f"{self.start_bold}-----------------------------------------------{self.end_bold}")
                return True
        except Exception as e:
            print(f"Failed to check_delta_breach.{e}")
            traceback.print_exc()

    def create_close_leg(self, position, typ):
        try:
            # return True
            data = {
            'quantity': abs(position['quantity']),
            'product': 'D',
            'validity': 'DAY',
            'price': 0,
            'tag': 'string',
            'instrument_token': position['instrument_token'],
            'order_type': 'MARKET',
            'transaction_type': self.map_order_type[typ],
            'disclosed_quantity': 0,
            'trigger_price': 0,
            'is_amo': False
            }
            # Send the POST request
            response = requests.post(self.url_place_position, json=data, headers=self.headers)
            if response.status_code == 200 and response.json()['data']:
                print(f"{position['tradingsymbol']} is {typ}d successfully!")
                return True
            else:
                print(f"Failed to {typ} {position['tradingsymbol']}")
                return False
        except Exception as e:
            print(f"Failed to {typ} {position['tradingsymbol']} due to exception:{str(e)}")
            traceback.print_exc()
            return False

    def download_instrument_token(self):
        '''Download the instrument token file from url. This is the one time activitity for a session'''
        try:
            if os.path.exists(self.instrument_token_zip_fl_nm):
                os.remove(self.instrument_token_zip_fl_nm)
            print(f"Removed existing file '{self.instrument_token_zip_fl_nm}'.")
            # Download the file
            response = requests.get(self.url_instrument_token, stream=True)
            if response.status_code == 200:
                with open(self.instrument_token_zip_fl_nm, "wb") as f:
                    for chunk in response.iter_content(1024):
                        f.write(chunk)
                print(f"Download complete! File saved as '{self.instrument_token_zip_fl_nm}'.")
            else:
                print(f"Error downloading file:{self.instrument_token_zip_fl_nm} {response.status_code}")
            # Extract the file
            with gzip.open(self.instrument_token_zip_fl_nm, 'rb') as f_in:
                # Open the output file for writing
                with open(self.instrument_token_csv_fl_nm, 'wb') as f_out:
                    # Write the decompressed content to the output file
                    f_out.write(f_in.read())
            # Save csv to pandas df
            return pd.read_csv(self.instrument_token_csv_fl_nm)
        except Exception as e:
            print("Failed to download instrument token zip file")
            traceback.print_exc()

    def get_instrument_token(self, option_type, delta):
        '''Fetch the instrument token for a given delta'''
        try:
            option_chain_stat, option_chain_df = self.get_option_chain()
            if not option_chain_stat:
                print(f"get_instrument_token: Failed to fetch option chain details")
                return False, '', ''
            # Match the delta & return the appropriate instrument token.
            matched_option_geeks_df = option_chain_df[(option_chain_df.option_type == option_type)
                                                    & (option_chain_df.delta < self.max_thresold)
                                                    & (option_chain_df.delta > self.min_thresold)][['delta','instrument_key', 'expiry', 'option_type']]
            instrument_df = self.instrument_token_df[['instrument_key', 'expiry', 'option_type']]
            #return matched_option_geeks_df, instrument_df
            matched_option_geeks_df = pd.merge(instrument_df, matched_option_geeks_df, how='inner', on=['instrument_key', 'expiry', 'option_type'])
            #return matched_option_geeks_df
            matched_delta_lst = matched_option_geeks_df.delta.values
            matched_delta = min(matched_delta_lst, key=lambda x: abs(x - delta))
            inst_token = matched_option_geeks_df[matched_option_geeks_df.delta == matched_delta].instrument_key.values[0]
            return True, inst_token, self.instrument_trading_symbol_map_dict[inst_token]
        except Exception as e:
            print(f'Failed to fetch the instrument token')
            traceback.print_exc()
            return False, '', ''

    def get_active_pos_geeks(self):
        '''Merge active positions with option geeks and return the dict of pos with delta max & min respetively'''
        try:
            option_geeks_df_stat, option_geeks_df = self.get_option_chain()
            if not option_geeks_df_stat:
                print('get_active_pos_geeks: Failed to get option chain')
                return False, '', '', ''
            active_pos_df_stat, current_pnl, active_pos_df = self.get_active_positions()
            if not active_pos_df_stat:
                print('get_active_pos_geeks: Failed to get active position')
                return False, '', '', ''
            active_pos_option_geeks_df = pd.merge(active_pos_df, option_geeks_df, left_on='instrument_token',right_on='instrument_key', how='inner')
            act_pos_delta_max = active_pos_option_geeks_df[active_pos_option_geeks_df.delta.max() == active_pos_option_geeks_df.delta].to_dict(orient='records')[0]
            act_pos_delta_min = active_pos_option_geeks_df[active_pos_option_geeks_df.delta.min() == active_pos_option_geeks_df.delta].to_dict(orient='records')[0]
            return True, current_pnl, act_pos_delta_max, act_pos_delta_min
        except Exception as e:
            print('Failed to get active positions geeks')
            return False, '', '', ''

    def get_active_positions(self):
        '''Gets all the active positions'''
        try:
            response = requests.get(self.url_positions, headers=self.headers)
            if response.status_code == 200 and response.json()['data']:
                tmp_act_pos = pd.DataFrame(response.json()['data'])
                pnl = int(tmp_act_pos['pnl'].sum())
                tmp_act_pos = tmp_act_pos[(tmp_act_pos.sell_price != 0.0) & (tmp_act_pos.buy_price == 0.0)]  # change it to sell price later.
                if tmp_act_pos.shape[0] != 0:
                    return True, pnl, tmp_act_pos
                else:
                    print('get_active_positions: Position df is empty!')
                    return False, '', ''
            else:
                print('get_active_positions: Failed to get current positions')
                return False, '', ''
        except Exception as e:
            print('get_active_positions: Failed to get current positions')
            traceback.print_exc()
            return False, '', ''

    def get_option_chain(self):
        '''This method return the option chain dict for a given expiry.
        '''
        try:
            call_options_geeks_lst = []
            put_options_geeks_lst = []
            params = {
                'instrument_key': f"{self.instrument_index}|{self.instrument_key_map_dict[self.index]}",
                'expiry_date': self.expiry_date
                }

            response = requests.get(self.url_option_chain, params=params, headers=self.headers)
            if response.status_code == 200 and response.json()['data']:
                for i in response.json()['data']:
                    strike_price = int(i['strike_price'])
                    call_options_geeks = i['call_options']['option_greeks']
                    put_options_geeks = i['put_options']['option_greeks']
                    call_options_geeks['option_type'] = 'CE'
                    put_options_geeks['option_type'] = 'PE'
                    call_options_geeks['instrument_key'] = i['call_options']['instrument_key']
                    put_options_geeks['instrument_key'] = i['call_options']['instrument_key']
                    call_options_geeks['strike_option'] = str(strike_price)+'CE'
                    put_options_geeks['strike_option'] = str(strike_price)+'PE'
                    call_options_geeks['expiry'] = i['expiry']
                    put_options_geeks['expiry'] = i['expiry']
                    call_options_geeks['underlying_key'] = i['underlying_key']
                    put_options_geeks['underlying_key'] = i['underlying_key']
                    call_options_geeks['strike_price'] = strike_price
                    put_options_geeks['strike_price'] = strike_price
                    call_options_geeks_lst.append(call_options_geeks)
                    put_options_geeks_lst.append(put_options_geeks)
                option_geeks_df = pd.DataFrame(call_options_geeks_lst + put_options_geeks_lst)
                option_geeks_df['delta'] = option_geeks_df['delta'].apply(lambda x: abs(x))
                return True, option_geeks_df
            else:
                print('Failed to get the option chain!')
                return False, ''
        except Exception as e:
            print('Failed to get the option chain!')
            traceback.print_exc()
            return False, ''
```
